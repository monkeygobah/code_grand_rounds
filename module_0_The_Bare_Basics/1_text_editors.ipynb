# **Text editors and IDEs**

## **Intro**

### One of the most common questions people new to coding ask is where do you even go to write code on your computer.  When you go to write essays, you go to Word, and when you need to make a slideshow you go to powerpoint, so then where do you go to write a script to analyze your clinical data?

### The answer is text editors or  Integrated Development Environments (IDEs). Text editors and IDEs are platforms, like Microsoft word and PowerPoint that are built to facilitate developing software. There are technical differences between text editors and IDEs (IDEs provide more features to help code but are heavier and more complex), but will basically treat them as interchangeable here for the sake of simplicity.

### There are many different text editors that focus on different things (various coding languages, efficiency, etc), but at their core they are all designed to meet the needs of people writing code and they all have similar features. 

## **VSCode**

### To get started writing code, we recommend you use VSCode. VSCode is a multipurpose text editor that will allow you to write code in any computer language (including python which is the focus of code grand rounds). It also has support for various add ons and other tricks which make your life easier and make it basically a very lightweight IDE. As you progress through yout coding career you may decide to switch to other ones (PyCharm and VIM are also very popular ones), but VSCode is the best place to start. To download, go to this link (https://code.visualstudio.com/download) and install it based on your spcific machine requirements.

### Once you go through the process and launch the app on your computer, you will get to the home page of VSCode. 

### You can import both individual files (for python or Jupyter notebooks, these are files that end in .py or .ipynb respectively) or folders (where your code and data might be stored on your computer) to work on. If you download files from Code Grand Rounds, you might put the Jupyter notebook file and data in the same folder which you will load into VSCode (be careful of paths to files within the code- you may have to change some to match where the data lives on your machine. You will learn about this later in the course!).  You can drag and drop or use file>open file (or folder) within VSCode. You can also click `cmd + p` (on mac) to quick open files.

## **More Handy Features**

### One of the most useful things you can is to add multiple cursors. To do this you just `ctrl + click (on mac)` at multiple points on your screen. If you have to type the same thing at multiple places, which happens more often than not, this is an extremely helpful trick (make sure you press `esc` a few times to go back to normal editing).

### If you ever have to indent multiple lines of code, you can highlight all of them and click `tab` which all allow you to indent them all at the same time. To unindent multiple lines, highlight the lines of interest and click `shift + tab` to move all the lines out an indentation level. Indents and whitespace are very important in Python and you will see why in Module 1!

### The last efficiency trick we will share is the ability to `comment out` large blocks of code quickly (this just tells the computer to ignore these lines, again, more to come on this). You can `comment out` multiple lines at once by highlighting the lines of interest and clicking `cmd + /`. You can reverse the commenting the same way!

### This is just a brief overview of all the things VSCode can do, but it is more than enough to get you started. If you want to get the most out of VSCode it is worth watching a few YouTube videos and completing the tutorial that is on the app when you log in!


## **Note about Jupyter Notebooks and Google Colab**

### When you start to use Jupyter notebooks, you will notice that in addition to `Python` cells (cells are the little blocks that you write into) you will also see `Markdown` cells. `Markdown` is a markup language that allows you to write and quickly format text to explain your code in plain english. All of the not code stuff on Code Grand Rounds is written in `Markdown`. You don't have to use this feature, but we want to mention it so you know about it. There are many more helpful buttons in VSC designed for improving your experience using Jupytner Notebooks, but it just too much to cover here. We reccomend you watch a few you Youtube videos (https://www.youtube.com/watch?v=DA6ZAHBPF1U is a good one) to see the full breadth of what it has to offer.

### Google Colab is a cloud-based platform provided by Google that allows users to write and execute Python code in a web-based, interactive environment. It offers a similar experience to Jupyter Notebooks, but you can run it in the cloud and access GPUs (for free) which you might not otherwise have access to. Again, this is something you might not strictly need to know but you will hear the words so we wanted to mention it!

### For colored images, these values are typically broken down into channelscommonly Red, Green, and Blue (RGB). Each channel represents the intensity of that particular color in the pixel. For example, an RGB value of (255,0,0) represents a purely red pixel, while (0,255,0) is purely green.

Image taken from https://www.geeksforgeeks.org/matlab-rgb-image-representation/

## Colors

### We will start with more manual operations to mess with the colors and end with some built in methods from OpenCV

### As mentioned before, color channels are stored as a list of 3 numbers for the Red, Green, and Blue channels. Each number represents how much of each color is represented in the pixel with 0 being none and 255 being the most. Thus, [255, 0, 0] would be a completely red pixel, [0, 255, 0] would be green, and so on. Once all the channels are used, we can make all different colors. Let's see what happens when we switch around the colors.

### Specifically, we are going to rotate the values for each channels so that all of the values for Red will be set to the new Green values, all of the values for Green will be the new Blue values, and the Blue values will be the new Red values

```python
# Make a copy of Ben
ben_GBR = ben.copy()
# Save a temporary copy of the Red channel 
benR = ben[:, :, 0] 
# Set the R as G
ben_GBR[:, :, 0] = ben_GBR[:, :, 1]
# Set the G as B
ben_GBR[:, :, 1] = ben_GBR[:, :, 2]
# Set the B as old R
ben_GBR[:, :, 2] = benR
```

```python
# Show the new image (this would now be [Green, Blue, Red])
imgplot = plt.imshow(ben_GBR)
```

### It's interesting that you can still tell this is Ben with his friends, but all of the colors are weirdly messed up. Take a stab at practicing manually changing the color channel orientations and make BRG or any other combination before moving on!

```python
# Make a copy of Ben
ben_BGR = ben.copy()
# Save a temporary copy of the Red channel 
benR = ben[:, :, 0] 

## You do the rest
# Set the R as B
ben_BGR[:, :, 0] = ben_BGR[:, :, 2]
# Set the B as R
ben_BGR[:, :, 2] = benR

# display image
imgplot = plt.imshow(ben_BGR)
```

## Grayscale 

### Now, let's try putting on a "filter" such as a grayscale. As mentioned before, grayscale is represented using only one color channel, so we will need to consolidate the information from three color channels into one. How do we do that? There are two ways to calculate grayscale using the existing color channels manually:

1. $\frac{(R+G+B)}{3}$
2. $ 0.299\cdot R + 0.587 \cdot G + 0.114 \cdot B$

```python
# Option 1
ben_gray = (ben[:, :, 0] + ben[:, :, 1] + ben[:, :, 2])/3
imgplot = plt.imshow(ben_gray, cmap='gray')
```

```python
# Option 2
ben_gray_weighted = 0.299*ben[:, :, 0] + 0.587*ben[:, :, 1] + 0.114*ben[:, :, 2]
imgplot = plt.imshow(ben_gray_weighted, cmap='gray')
```

### Why did option 2 look so much better? Because option 1 assumed each color channel (RGB) had equal importance to the final image makeup. Option 2 takes into account the relative contribution of each channel using luminance coefficients. You will *never* need to know how or why these works, but this is why the second option looks so much cleaner. Now that you have done everything manually, let's move on to a library that will make your lives a lot easier.

Let's quickly take a look through the multitude of options that we can choose from for colors. First, we have to reload being using the OpenCV tools. When OpenCV does this, it still stores the images as numpy arrays so all of the cropping functions we learned above still work. Let's import the image now.

### Oh NO! Why is ben blue... but wait... have you seen this before? Yes you have! 

### OpenCV loads images in the BGR channel format by default. This means that the all the red pixels are represented as blue and vice versa. We can easily fix this by using the convert color function from OpenCV `cvtColor`. This is pretty self explanatory, but basically this method takes BGR and converts to RGB

```python
gray_ben = cv2.cvtColor(ben_rgb, cv2.COLOR_RGB2GRAY) # note we use RGB now instead of BGR 
imgplot = plt.imshow(gray_ben, cmap='gray')
```

## Resizing

### Resizing images is a critical part of image processing. From a practical stance, medical images that we need for our research or want to analyze may be different sizes which makes it hard to write code to perform operations on them. By resizing, we can standardize the size of all of our images. These types of operations become particularly important for data preparation for machine learning. Fortunately this is very easy in OpenCV, because there is already a `resize` function. All you need is to give it a desired width and height.

```python
# Play around with these and see what happens
width = 700
height = 300

# Resize the image to the specified width and height
resized_image = cv2.resize(ben_rgb, (width, height))
imgplot = plt.imshow(resized_image)
```

### It would be nice to be able to resize an image without completely ruining it if the dimensions are off. How can we do this? There are two ways to go about it: using a scale factor and setting an aspect ratio. 

### A scale factor is exactly what it sounds like. It is a numerical value by which we will multiply the width and height of the image. This is really usefully when you have a lot of images that you want to make bigger or smaller, because you can set a single scale factor and then resize all of the images by that scale factor. The `resize` function can actually be used with a scale factor for both x and y (`fx` and `fy` respectively). Note the axes when the image is displayed

```python
# Set the scale factor for resizing
scale_factor = 1.5

print("Original shape:", ben_rgb.shape)

# Resize the image based on the scale factor
resized_image = cv2.resize(ben_rgb, None, fx=scale_factor, fy=scale_factor)
imgplot = plt.imshow(resized_image)
print("New shape:", resized_image.shape)
```

### The aspect ratio of an image is the ratio of the width to the height of the image. In order to keep the image from looking distorted, this ratio needs to be maintains when resizing the image! This is also very use if you have a lot of images of different sizes that you want to make the same size. The caveat is that you may need to pad (aka add on) additional pixels to make all of the images' aspect ratios match.

### In order to use this, we start off with one dimension that we know for sure we want. Then, we can calculate the aspect ratio of the original image. Using this aspect ratio, we can calculate the desired height to maintain the aspect ratio

```python
# Set the desired width while preserving the aspect ratio
new_width = 800

# Calculate the new height based on the aspect ratio
aspect_ratio = ben_rgb.shape[1] / ben_rgb.shape[0]
new_height = int(new_width / aspect_ratio)

# Resize the image to the new width and height
resized_image = cv2.resize(ben_rgb, (new_width, new_height))
imgplot = plt.imshow(resized_image)

print("Original shape:", ben_rgb.shape)
print("New shape:", resized_image.shape)
```

## Thresholding

### Sometimes we want to easily segment an image into meaningful regions based on some type of apriori knowledge about the data to simplify downstream usecases. Thresholding is a way to do that. We will now go through some of the thresholding options easily handed to you in openCV. We will use the grayscale image becasue each pixel only has one value. 

### To begin basic binary thresholding, we set a threshold or cutoff where any value below the cutoff will be set to black (`0`) and any value above the cutoff will be set to white (`255`). (As an FYI, if it is on the cutoff, it will be set to black)


Image taken from https://scikit-image.org/docs/stable/auto_examples/segmentation/plot_thresholding.html

```python
# first make sure we still have our RGB image of Ben to work with as a sanity check
imgplot = plt.imshow(gray_ben, cmap='gray')
```

```python
_, binary_image = cv2.threshold(gray_ben, 127, 255, cv2.THRESH_BINARY)
# The threshold in this case is 127.
imgplot = plt.imshow(binary_image,cmap='gray')
```

### There are also other types of thresholding like inverse binary thresholding which is basically the same function as the one seen except the black and white is reversed. 

### In adaptive thresholding, the threshold value is calculated using the mean of each neighborhood. The "neighborhoods" of the image is small regions (11x11) and the threshold is calculated for each region based on the mean value of that region.

### In a similar vein, in adaptive Gaussian thresholding, the threshold is calculated by performing a weighted sum of the neighborhood values, where the weights are based on a Gaussian window. The image is divided into small regions (11x11) and the threshold is calculated for each region based on the weighted sum.

### Finally, there is Otsu's thresholding which is an algorithm that goes beyond what you need to know, but the threshold separates the image into foreground and background based on minimizing the variance between the classes.

```python
_, binary_inv_image = cv2.threshold(gray_ben, 127, 255, cv2.THRESH_BINARY_INV)
imgplot = plt.imshow(binary_inv_image,cmap='gray')
```

```python
adaptive_mean_image = cv2.adaptiveThreshold(gray_ben, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
imgplot = plt.imshow(adaptive_mean_image,cmap='gray')
```

```python
adaptive_gaussian_image = cv2.adaptiveThreshold(gray_ben, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
imgplot = plt.imshow(adaptive_gaussian_image,cmap='gray')
```

```python
_, otsu_image = cv2.threshold(gray_ben, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
imgplot = plt.imshow(otsu_image,cmap='gray')
```

## Blurring

### Blurring is a fundamental technique to remove noise from images. By noise, we mean the little details and variance in images that can be confusing to machine learning models and analyses. Sometimes, we don't want the computer to be focusing on these small details. The math behind how these algorithms work can be complex, but we will now demonstrate some of the built in OpenCV methods to blur/smooth an image

```python
# Apply Gaussian blur to the image
# arguments: image, kernel, standard deviation (0 calculates automatically. Use 0 unless you have a good reason not to)
gaussian_blur_image = cv2.GaussianBlur(ben_rgb, (5, 5), 0) 
imgplot = plt.imshow(gaussian_blur_image)
```

### Hm... the image didn't change that much. This is because we set the kernel too small. The kernel is basically the size of the area around which each pocket is blurred. Let's increase the size of the kernel to increase the blur.

```python
# Apply Gaussian blur to the image
# arguments: image, kernel (must be POSITIVE ODD integers), 
# standard deviation (0 calculates automatically. Use 0 unless you have a good reason not to)
gaussian_blur_image = cv2.GaussianBlur(ben_rgb, (205, 205), 0) 
imgplot = plt.imshow(gaussian_blur_image)
```

```python
# Apply median blur to the image
# arguments: image, number of neighbors for each pixel to take median value of (must be positive odd integer)
median_blur_image = cv2.medianBlur(ben_rgb, 105) 
imgplot = plt.imshow(median_blur_image)
```

## Edge Detection

### Edge detection is the most complicated methods we will talk about here, but also one of the most powerful. It can be used to to automatically find edges in images. It is not hard to think of many use cases. Do not worry at this point about understanding the underlying algorithms, but instead know that these agorothms exist and you can apply them to your problems. Here, we will go through two such algorithms: Canny and Sobel's edge detection.

```python
### Apply Canny edge detection to gray ben

# set some intial thresholds and tweak to achieve desired result. This requires lots of 'playing'...
# lower threshold- any gradient value below this value is ignored
threshold1 = 100
# upper threshld- edges with gradient value above this threshold are strong and kept
threshold2= 230


edges = cv2.Canny(gray_ben, threshold1, threshold2)

# Display the edges
imgplot = plt.imshow(edges, cmap='gray')
```

```python
# Apply Sobel edge detection to gray ben

# Determine the size of the Sobel kernel
sobel_kernel = 3

# Set up the Sobel operator to detect edges

# Apply the Sobel operator along the x-axis (horizontal gradient) using cv2.Sobel()
sobel_x = cv2.Sobel(gray_ben, cv2.CV_64F, 1, 0, ksize=sobel_kernel)

# Apply the Sobel operator along the y-axis (vertical gradient) using cv2.Sobel()
sobel_y = cv2.Sobel(gray_ben, cv2.CV_64F, 0, 1, ksize=sobel_kernel)

# Calculate the magnitude of the edge gradients

# Calculate the magnitude of the gradients using numpy.sqrt() by combining the squared x and y gradients
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Display the edge magnitude

# Display the magnitude image using plt.imshow() with the 'gray' colormap
imgplot = plt.imshow(magnitude, cmap='gray')

```

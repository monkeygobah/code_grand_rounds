# Coding Applications in Medicine: V/Q Scans

### Ventilation-perfusion scans (V/Q) are a very useful tool in medicine. It is a special imaging modality that allows doctors to monitor the air and blood that go through the lungs. By comparing the space air and blood takes up in the lung, we can determine if the patient has lung pathologies such as COPD or, in extreme cases, a pulmonary embolism. The ventilation and perfusion scans are taken separately, so we will have figure out how to compare the images using code in order to diagnose the patient. Let's take a look how we can do that using the toolset we built from the computer vision basics notebook.

```python
# Import the relevant packages
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
import os
from PIL import Image
```

## Set Up

### Throughout this project, we will be working with 3 sets of images: a normal lung V/Q, a COPD lung V/Q, and a pulmonary embolism lung V/Q. We will start by looking at a normal lung scan to see what this imaging modality looks like and the images we're going to be working with. 

```python
# Import the images of a normal lung V (ventilation) and Q (perfusion) scan as greyescale
normal_V = cv.imread('VQ_images/normal_V.png', cv.IMREAD_GRAYSCALE)
normal_Q = cv.imread('VQ_images/normal_Q.png', cv.IMREAD_GRAYSCALE)
```

```python
# Let's look at the image that we imported
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(normal_V, cmap='gray')
ax2.imshow(normal_Q, cmap='gray')
ax1.set_title('Normal V')
ax2.set_title('Normal Q')
```

## Thresholding

### Next we're going to threshold the image. Because the image is on the grayscale, the value of each pixel is between 0 and 255. Thresholding sets each pixel of the image to one of the extreme (black or white). Thankfully, the OpenCV package has one for us to use. It takes an image and a threshold value. Then, it sets everything above the threshold value to 255 (white) and everything below the threshold to 0 (black) due to cv.THRESH_BINARY_INV (the type of thresholding the function is using). https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html has more types if you want to explore!

### By thresholding the image, we can isolate the important part of the image: the areas of the lung that are well perfused or well ventilated. Let's play around with different threshold values to see which ones isolate the area that we want.

```python
# Play around with the threshold value 0-255 and see what happens
thresh_val = [75, 100, 125, 150, 175]

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(normal_V, cmap='gray')
ax2.imshow(normal_Q, cmap='gray')
ax1.set_title('Normal V')
ax2.set_title('Normal Q')

for thresh in thresh_val:
    _, normal_V_thresh = cv.threshold(normal_V, thresh, 255, cv.THRESH_BINARY_INV)
    _, normal_Q_thresh = cv.threshold(normal_Q, thresh, 255, cv.THRESH_BINARY_INV)
    
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.imshow(normal_V_thresh, cmap='gray')
    ax2.imshow(normal_Q_thresh, cmap='gray')
    ax1.set_title('Threshold = '+ str(thresh))
```

### It looks like the best threshold value is between 125 and 150. These values give us all of the lung unlike 75 and 100, but also compared to 175, it doesn't pick up on extraneous information. We call that extraneous information noise. Let's go with 125, because this value preserves the important information while also filtering out a good amount of noise.

## Binary Operations

### While we see the values of the each pixel as 0-255, the computer actually stores each pixel value using a number system called binary numbers. This system represents all numbers as a combination of only 1's and 0's, which are called bits. In our case, white is represented by all 1's (11111111) and black is represented by all 0's (00000000)

### We can use this to our advantage with operations on the bits. Normal arithmetic operations like addition and subtraction still work, but now we have unique operations called AND and OR.

### 0 AND 0 = 0, 0 AND 1 = 0, 1 AND 1 = 1
### 0 OR 0 = 0, 0 OR 1 = 1, 1 OR 1 = 1

### See a pattern?

### This is more than you will need to know, but it is a very helpful operation when we are comparing the images. We can actually think of this as the intersection and union between the images. When we use the AND function, it will only preserve the overlapping white pixels that exist on BOTH images. When we use the OR function, it will preserve the white pixels as long as it existed on one of the two original images. Let's see how we can use this to our advantage when comparing the two images.

```python
# Perform the AND
_, normal_V_thresh = cv.threshold(normal_V, 125, 255, cv.THRESH_BINARY_INV)
_, normal_Q_thresh = cv.threshold(normal_Q, 125, 255, cv.THRESH_BINARY_INV)
VQ_and = cv.bitwise_and(normal_V_thresh, normal_Q_thresh)
```

```python
# Let's see what it did
plt.imshow(VQ_and, cmap='gray')
```

### Once we performed the AND operation on our ventilation and perfusion threshholded images, we can find the intersection of the two. Why is this useful to us? This actually tells use which parts of the lung are BOTH well-perfusion and well-ventilated. Thus, if there is an area that is not getting enough air or blood, we will be able to see that. 

### Can we looking at this from a more quantitative perspective? Yes! Let's try to count what percent of each lung is both well-perfused AND well-ventilated.

```python
# Count how many white pixels there are in the intersection of the two thresholded images
mask_pixels = np.count_nonzero(VQ_and == 255)

# Count how many white pixels there are in each thresholded images
normal_V_pixels = np.count_nonzero(normal_V_thresh == 255)
normal_Q_pixels = np.count_nonzero(normal_Q_thresh == 255)

# Calculate the percent of the pixels that are well-perfused and well-ventilated
print('Percent of Ventilation covered: ', mask_pixels/normal_V_pixels)
print('Area of perfusion that is also ventilated: ', mask_pixels/normal_Q_pixels)
```

## Create a mask

### Now that we know which areas of the lung are both well-ventilated and well-perfused, how can we use this information to figure out which areas in the lung are NOT well-ventilated or well-perfused. Well, we can actually superimpose this intersection we created onto the original images. This is called creating a mask which we can put on the original images. In order to make it more visible, we will make the mask red.

```python
# First convert the image to RGB (3 color channels)
V_to_rgb = cv.cvtColor(normal_V, cv.COLOR_GRAY2RGB)
Q_to_rgb = cv.cvtColor(normal_Q, cv.COLOR_GRAY2RGB)
```

```python
# Create an image that's all black
red_mask = np.zeros((normal_V.shape[0], normal_V.shape[1],3), dtype='uint8')

# and make the Red color channel the same as the thresholded image
red_mask[:,:,0] = VQ_and
```

```python
# Now we have a red image!
plt.imshow(red_mask)
```

### Once we have this mask, we have to add it on top of the original image. There is a helpful function in OpenCV called `cv.addWeighted` which allows you to combine two images. The weights allows you to choose which image you want to be more prominent/see better. In essence we are blending the two images together. An example of this is shown below.

Image taken from https://opencv24-python-tutorials.readthedocs.io/en/latest/py_tutorials/py_core/py_image_arithmetics/py_image_arithmetics.html

```python
# Time to add this red mask on top of the original images. The weight controls how opaque we want the red
# on top of the original image. Play around with this value!
weight = 0.7
combined_V = cv.addWeighted(V_to_rgb, weight, red_mask, 1-weight, 0)
combined_Q = cv.addWeighted(Q_to_rgb, weight, red_mask, 1-weight, 0)
```

```python
# Show the final image
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.imshow(combined_V)
ax2.imshow(combined_Q)
ax1.set_title('V with Mask')
ax2.set_title('Q with Mask')
```

## Next Steps

### Obviously, this was not the most ideal mask since a lot of the lung was not covered. But, now that we have done this with a normal lung, you can perform the same operations for the COPD lung and pulmonary embolism lung. Below, we created a function that allows you to go through the entire process with one call. 

### The options for the function are based of the image set that you want to use. Input `'normal'`, `'copd'`, or `'pe'` as a string intot he function and it will run the process on that set of images (case sensitive). Try to look at the similarities and differences! 

### You can also mess around with the functions in the algorithm like the thresholding values or the type of threshholding to see if you can create a better mask. If you want an extra challenge, try to also use the edge detection code and try to outline the lungs!

```python
# normal, copd, or pe
def create_mask(image_set):
    # Import the images of a normal lung V (ventilation) and Q (perfusion) scan as greyescale
    V = cv.imread('VQ_images/' + image_set + '_V.png', cv.IMREAD_GRAYSCALE)
    Q = cv.imread('VQ_images/' + image_set + '_Q.png', cv.IMREAD_GRAYSCALE)
    
    # Let's look at the image that we imported
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.imshow(normal_V, cmap='gray')
    ax2.imshow(normal_Q, cmap='gray')
    ax1.set_title(image_set + ' V')
    ax2.set_title(image_set + ' Q')
    
    # Threshold the images
    _, V_thresh = cv.threshold(V, 125, 255, cv.THRESH_BINARY_INV)
    _, Q_thresh = cv.threshold(Q, 125, 255, cv.THRESH_BINARY_INV)
    
    # Perform the AND
    VQ_and = cv.bitwise_and(V_thresh, Q_thresh)
    
    # Count how many white pixels there are in the intersection of the two thresholded images
    mask_pixels = np.count_nonzero(VQ_and == 255)

    # Count how many white pixels there are in each thresholded images
    V_pixels = np.count_nonzero(V_thresh == 255)
    Q_pixels = np.count_nonzero(Q_thresh == 255)

    # Calculate the percent of the pixels that are well-perfused and well-ventilated
    print('Percent of Ventilation covered: ', mask_pixels/V_pixels)
    print('Area of perfusion that is also ventilated: ', mask_pixels/Q_pixels)
    
    # Convert the image to RGB (3 color channels)
    V_to_rgb = cv.cvtColor(V, cv.COLOR_GRAY2RGB)
    Q_to_rgb = cv.cvtColor(Q, cv.COLOR_GRAY2RGB)
    
    # Create an image that's all black
    red_mask = np.zeros((V.shape[0], V.shape[1],3), dtype='uint8')

    # and make the Red color channel the same as the thresholded image
    red_mask[:,:,0] = VQ_and
    
    # Add the mask onto the original images
    weight = 0.7
    combined_V = cv.addWeighted(V_to_rgb, weight, red_mask, 1-weight, 0)
    combined_Q = cv.addWeighted(Q_to_rgb, weight, red_mask, 1-weight, 0)
    
    # Show the final image
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.imshow(combined_V)
    ax2.imshow(combined_Q)
    ax1.set_title('V with Mask')
    ax2.set_title('Q with Mask')
```

```python
# Run the function that was created above
create_mask('copd')
```

```python
# Run the function that was created above
create_mask('pe')
```
